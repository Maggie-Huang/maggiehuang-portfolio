# 1. Get Channel Statistics

```python
# retrieve channel data from YouTube using the channel ID
request = youtube.channels().list(
    part='snippet,contentDetails,statistics',
    id=channel_id
)
response_channel = request.execute()

response_channel
```

```python
# fetch channel statistics from the response_channel dictionary
channel_stats = dict(
    Title = response_channel['items'][0]['snippet']['title'],
    Description = response_channel['items'][0]['snippet']['description'],
    URL = 'https://www.youtube.com/' + response_channel['items'][0]['snippet']['customUrl'],
    Country = response_channel['items'][0]['snippet']['country'],
    Views = int(response_channel['items'][0]['statistics']['viewCount']),
    Subscribers = int(response_channel['items'][0]['statistics']['subscriberCount']),
    Videos = int(response_channel['items'][0]['statistics']['videoCount'])
)

channel_stats
```

# 2. Get video IDs through upload ID

## 2.1. Get upload ID

```python
# fetch uploads from response_channel and assign it to upload_id
upload_id = response_channel['items'][0]['contentDetails']['relatedPlaylists']['uploads']

upload_id
```

## 2.2. Get video IDs

```python
# retrieve the first 50 video items using the upload ID
request = youtube.playlistItems().list(
    part='contentDetails',
    playlistId = upload_id,
    maxResults = 50
)

response_videos = request.execute()
response_videos
```

```python
# create a list named video_ids
video_ids = list()

# fetch video IDs and append them to the list
for item in response_videos['items']:
    video_ids.append(item['contentDetails']['videoId'])

video_ids[0:5]
```

```python
# fetch nextPageToken from response_videos and assign it to next_page_token
next_page_token = response_videos['nextPageToken']

next_page_token
```

```python
# retrieve video items using the upload ID and next page token, fetch video IDs and append them to the list 
more_pages = True
while more_pages:
    request = youtube.playlistItems().list(
        part='contentDetails',
        playlistId = upload_id,
        maxResults = 50,
        pageToken = next_page_token
    )
    response_videos = request.execute()
    
    for item in response_videos['items']:
        video_ids.append(item['contentDetails']['videoId'])
    
    try:
        next_page_token = response_videos['nextPageToken']
    except KeyError:
        more_pages = False

len(video_ids)
```

# 3. Get video data and store them in Pandas DataFrame

```python
# create a Pandas DataFrame with column titles
df = pd.DataFrame(
    columns=[
        'ID', 'Title', 'Published At', 'Category', 'Default Audio Language', 'Duration', 'Definition', 'Caption',
        'Licensed', 'View Count', 'Like Count', 'Dislike Count', 'Favorite Count', 'Comment Count', 'Description'
    ]
)
```

```python
# retrieve video details using the video ID, fetch the information required, and store them to the DataFrame
for ID in video_ids:

    # retrieve video details using video ID
    request = youtube.videos().list(
        part='snippet,contentDetails,statistics',
        id=ID
    )
    response_video = request.execute()

    # fetch data from response
    published = response_video['items'][0]['snippet']['publishedAt']
    title = response_video['items'][0]['snippet']['title']

    try:
        description = response_video['items'][0]['snippet']['description']
    except KeyError:
        description = np.nan

    try:
        category = response_video['items'][0]['snippet']['categoryId']
    except KeyError:
        category = np.nan
    
    try:
        audio = response_video['items'][0]['snippet']['defaultAudioLanguage']
    except KeyError:
        audio = np.nan

    try:
        duration = response_video['items'][0]['contentDetails']['duration']
    except KeyError:
        duration = np.nan

    try:
        definition = response_video['items'][0]['contentDetails']['definition']
    except KeyError:
        definition = np.nan

    try:
        caption= response_video['items'][0]['contentDetails']['caption']
    except KeyError:
        caption = False

    try:
        licensed = response_video['items'][0]['contentDetails']['licensedContent']
    except KeyError:
        licensed = False

    try:
        view = int(response_video['items'][0]['statistics']['viewCount'])
    except KeyError:
        view = 0

    try:
        favorite = int(response_video['items'][0]['statistics']['favoriteCount'])
    except KeyError:
        favorite = 0

    try:
        like = int(response_video['items'][0]['statistics']['likeCount'])
    except KeyError:
        like = 0
    
    try:
        dislike = int(response_video['items'][0]['statistics']['dislikeCount'])
    except KeyError:
        dislike = 0

    try:
        comment = int(response_video['items'][0]['statistics']['commentCount'])
    except KeyError:
        comment = 0

    # store fetched data in a list
    row_data = [
        ID, title, published, category, audio, duration, definition, caption, licensed, view, like, dislike, favorite, comment, description
    ]

    # insert new row of data to the DataFrame
    length = len(df)
    df.loc[length] = row_data

df
```
